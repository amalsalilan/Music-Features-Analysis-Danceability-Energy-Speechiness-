# EDA

```python
# Summary statistics
summary_stats = df.describe()
print(summary_stats)

# Histogram of a numerical variable
plt.hist(df['Duration_ms'], bins=20)
plt.xlabel('Duration (ms)')
plt.ylabel('Frequency')
plt.title('Distribution of Song Durations')
plt.show()

# Box plot of a numerical variable
plt.boxplot(df['Energy'])
plt.ylabel('Energy')
plt.title('Box Plot of Energy')
plt.show()

# Scatter plot of two numerical variables
plt.scatter(df['Danceability'], df['Valence'])
plt.xlabel('Danceability')
plt.ylabel('Valence')
plt.title('Relationship between Danceability and Valence')
plt.show()

# Correlation matrix
corr_matrix = df.corr()
plt.imshow(corr_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.title('Correlation Matrix')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = df
# Most viewed videos
most_viewed_videos = data.sort_values('Views', ascending=False)[:5]
print("Most viewed videos:")
print(most_viewed_videos[['Title', 'Views']])
```

```python
# Visualize most viewed videos
plt.figure(figsize=(10, 6))
plt.barh(most_viewed_videos['Title'], most_viewed_videos['Views'])
plt.xlabel('Views')
plt.ylabel('Video Title')
plt.title('Most Viewed Videos')
plt.show()
```

```python
# Average likes per video
average_likes_per_video = data['Likes'].mean()
print("\nAverage likes per video:", average_likes_per_video)
```

```python
# Most liked videos
most_liked_videos = data.sort_values('Likes', ascending=False)[:5]
print("\nMost liked videos:")
print(most_liked_videos[['Title', 'Likes']])
```

```python
# Visualize most liked videos
plt.figure(figsize=(10, 6))
plt.barh(most_liked_videos['Title'], most_liked_videos['Likes'])
plt.xlabel('Likes')
plt.ylabel('Video Title')
plt.title('Most Liked Videos')
plt.show()

# Average comments per video
average_comments_per_video = data['Comments'].mean()
print("\nAverage comments per video:", average_comments_per_video)
```

```python
# Most commented videos
most_commented_videos = data.sort_values('Comments', ascending=False)[:5]
print("\nMost commented videos:")
print(most_commented_videos[['Title', 'Comments']])
```

```python
# Visualize most commented videos
plt.figure(figsize=(10, 6))
plt.barh(most_commented_videos['Title'], most_commented_videos['Comments'])
plt.xlabel('Comments')
plt.ylabel('Video Title')
plt.title('Most Commented Videos')
plt.show()
```

```python
# Most active channels based on comments
most_active_channels = data.groupby('Channel')['Comments'].sum().nlargest(5)
print("\nMost active channels based on comments:")
print(most_active_channels)
```

```python
# Visualize most active channels based on comments
plt.figure(figsize=(10, 6))
most_active_channels.plot(kind='bar')
plt.xlabel('Channel')
plt.ylabel('Total Comments')
plt.title('Most Active Channels Based on Comments')
plt.show()
```

```python
# Engagement rate
data['Engagement_rate'] = (data['Likes'] + data['Comments']) / data['Views']
print("\nEngagement rate:")
print(data[['Title', 'Engagement_rate']])
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example dataset with music information
dataset = pd.DataFrame({
    'Track': ['Track 1', 'Track 2', 'Track 3', 'Track 4', 'Track 5'],
    'Genre': ['Pop', 'Hip-Hop', 'Rock', 'Pop', 'Rock'],
    'Streams': [5000, 3000, 2000, 6000, 4000],
    'Likes': [100, 50, 30, 80, 60],
    'Comments': [10, 5, 2, 8, 6]
})

# Group the dataset by genre and calculate genre-specific metrics
genre_stats = dataset.groupby('Genre').agg({
    'Track': 'count',
    'Streams': 'sum',
    'Likes': 'sum',
    'Comments': 'sum'
}).reset_index()

# Sort the dataset by the number of tracks in each genre
genre_stats = genre_stats.sort_values(by='Track', ascending=False)

# Create advanced visualizations for genre analysis
plt.figure(figsize=(12, 6))

# Plot the number of tracks per genre
plt.subplot(2, 2, 1)
sns.barplot(x='Genre', y='Track', data=genre_stats, palette='muted')
plt.title('Number of Tracks per Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Tracks')

# Plot the total streams per genre
plt.subplot(2, 2, 2)
sns.barplot(x='Genre', y='Streams', data=genre_stats, palette='muted')
plt.title('Total Streams per Genre')
plt.xlabel('Genre')
plt.ylabel('Total Streams')

# Plot the total likes per genre
plt.subplot(2, 2, 3)
sns.barplot(x='Genre', y='Likes', data=genre_stats, palette='muted')
plt.title('Total Likes per Genre')
plt.xlabel('Genre')
plt.ylabel('Total Likes')

# Plot the total comments per genre
plt.subplot(2, 2, 4)
sns.barplot(x='Genre', y='Comments', data=genre_stats, palette='muted')
plt.title('Total Comments per Genre')
plt.xlabel('Genre')
plt.ylabel('Total Comments')

plt.tight_layout()
plt.show()
```
